```python
## DIY畫 多啦A夢
import numpy as np
import matplotlib.pyplot as plt
π = np.pi

## 臉和外框
θ=np.linspace(0,2*π,500)
β=np.linspace(-1.28*π,0.28*π,500)
α=np.linspace(-1.05*π,0.05*π,500)

r=3
x=r*np.cos(θ)
y=r*np.sin(θ)
d=2.5
fx=d*np.cos(β)
fy=-0.3+d*np.sin(β)
plt.plot(x,y,"B")
plt.plot(fx,fy,"B")

## 兩眼
e=0.8
Erx=0.8+e*np.cos(θ)
Ery=1.7+e*np.sin(θ)
Elx=-0.8+e*np.cos(θ)
Ely=1.7+e*np.sin(θ)
plt.plot(Erx,Ery,"B")
plt.plot(Elx,Ely,"B")

e1=0.05

for i in range(3,10):
      Erxx=0.8+i*e1*np.cos(θ)
      Eryy=1.7+i*e1*np.sin(θ)
      Elxx=-0.8+i*e1*np.cos(θ)
      Elyy=1.7+i*e1*np.sin(θ)
      plt.plot(Erxx,Eryy,"B")
      plt.plot(Elxx,Elyy,"B")  

## 鼻子
n=0.45
Nx=n*np.cos(θ)
Ny=0.3+n*np.sin(θ)
plt.plot(Nx,Ny,"B")

## 嘴吧
m=1.8
mx=m*np.cos(α)
my=-0.8+m*np.sin(α)
plt.plot(mx,my,"B")

M1x=[-1.75, 1.75]
M1y=[-0.5,-0.5]
plt.plot(M1x,M1y,"B")

M3x=[0   , 0  ]
M3y=[-0.5,-0.2]
plt.plot(M3x,M3y,"B")

M2x=(m-0.6)*np.cos(-α)
M2y=-2+(m-0.7)*np.sin(-α)
plt.plot(M2x,M2y,"B")

## 鬍鬚
M5x=[0.5,2]
M5y=[0.5,1]
plt.plot(M5x,M5y,"B")

M6x=[-0.5,-2]
M6y=[ 0.5, 1]
plt.plot(M6x,M6y,"B")

M7x=[ 0.5, 2.2]
M7y=[ 0.2, 0.2]
plt.plot(M7x,M7y,"B")

M8x=[-0.5,-2.2]
M8y=[ 0.2, 0.2]
plt.plot(M8x,M8y,"B")

M9x=[ 0.4, 2.1]
M9y=[-0.1,-0.4]
plt.plot(M9x,M9y,"B")

M0x=[-0.4,-2.1]
M0y=[-0.1,-0.4]
plt.plot(M0x,M0y,"B")
```
